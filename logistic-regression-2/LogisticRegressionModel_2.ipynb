```python
col_list = []
for col in census_data1.columns:
    if ((census_data1[col].dtype == 'object') & (col != 'y') ):
        col_list.append(col)
```

```python
census_data1.drop(columns = "Annual Income1", inplace=True)
```

```python
col_list
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
for i in col_list:
    census_data1[i]=labelencoder.fit_transform(census_data1[i])
```

```python
x = census_data1.iloc[:,:-1]
```

```python
x.info()
```

```python
y = pd.DataFrame(census_data["annual_income"])
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticRegr = LogisticRegression()
```

```python
logisticRegr.fit(x_train, y_train)
```

```python
test_pred = logisticRegr.predict(x_test)
```

```python
pred = pd.DataFrame()
pred["y_test"] = y_test
pred["y_pred"] = test_pred
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,test_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
c1=confusion_matrix(y_test,test_pred)
```
